```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
df = pd.read_csv('data/diabetes.csv')
print(df.shape)
df.head()
```

```python
df_values = df.values
df_values[:3, :]
```

```python
X = df_values[:, :-1]
X[:3]
```

```python
Y = df_values[:, -1]
Y[:3]
```

```python
D_Train, d_test, Y_Train, y_test = train_test_split(X, Y, stratify = Y, test_size = 0.3, random_state = 6)
```

```python
d_train, d_cv, y_train, y_cv = train_test_split(D_Train, Y_Train, stratify = Y_Train, test_size = 0.3, random_state = 7)
```

```python
model = Sequential()
model.add(Dense(2, input_dim =8, activation = 'relu' ))
model.add(Dense(1, activation = 'sigmoid' ))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(d_train, y_train, epochs = 100, batch_size = 10, verbose = 2)
```

```python
loss, accuracy = model.evaluate(d_cv, y_cv)
print(loss)
print(accuracy)
```

```python
predictions = model.predict_classes(d_test)
predictions
```

```python
score = accuracy_score(predictions, y_test)
score
```
